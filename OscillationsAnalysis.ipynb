# Oscillations notebook
This notebook gives an example on how to look for oscillations in the preprocessed data.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.jet()
from struvana.analysis import OscillationAnalysis
```

`load_file` should point to the numpy archive produced by the `struvana.preprocess.RawDataManager` class. In particular it is suggested to analyze the file produced after the dechirping and smoothing. 

```python
load_file = "/Users/lorenzo/Documents/PhD/StratusUV/Experiments/uridine/processed_data/d220527_01_pseudo_294_magic_dechirp.npz"
```

```python
fourier = OscillationAnalysis(load_file)
```

Let's plot to check the file is the one we expect:

```python
fourier.plot_map(figsize=(10,7), mOD=True)
```

## Get oscillations 
Get oscillations subtracting the original map to a smoothed one.

The `delayRange` arg is used to zoom the plot in this delay range.

```python
fourier.get_oscillations(smoothWindowDelayFS=150, delayRange=(200,1000), figsize=(10,7), mOD=True)
```

```python
fourier.get_oscillations_svd(2, delayRange=(-10,1000), figsize=(10,7), mOD=True)
```

## Fourier analysis
Let's compute the fourier transform between the delays 200 and 600:

```python
fourier.fourier_transform(200, 800)
```
